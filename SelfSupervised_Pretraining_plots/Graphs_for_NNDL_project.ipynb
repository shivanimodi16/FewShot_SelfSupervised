```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
plt.style.use('seaborn-whitegrid')
```

```python
file1 = open('train_losses_0.03.txt', 'r')
Lines = file1.readlines()
print(len(Lines))
```

```python
train_losses_first = []
for line in Lines:
    split_lines = line.split(":")
    train_losses_first.append(float(split_lines[1]))

print(len(train_losses_first))
```

```python
file2 = open('test_accuracies_linear_0.03.txt', 'r')
Lines2 = file2.readlines()
print(len(Lines2))
```

```python
test_losses_first = []
test_accuracies_first=[]
for line in Lines2:
    if('Train' in line):
      split_lines = line.split(":")
      #print("split_lines", split_lines)
      test_losses_first.append(float(split_lines[1].split(',')[0]))
      test_accuracies_first.append(float(split_lines[2]))
#print(len(test_losses_first))
```
